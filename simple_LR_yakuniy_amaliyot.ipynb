```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/reklama_sales.csv')
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df, x='TV', y='Sales', line_kws={"color": "red"})
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df, x='Radio', y='Sales', line_kws={"color": "red"})
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df, x='Newspaper', y='Sales', line_kws={"color": "red"})
plt.show()
```

```python
X = df['TV'] # TV ustuni Sales bilan bog'liqligi kuchli ekan
X.head()
```

```python
y = df['Sales']
y.head()
```

image.png
image.png

```python
xmean = X.mean()
ymean = y.mean()
```

```python
theta1 = sum((X-xmean)*(y-ymean))/sum((X-xmean)**2)
theta1
```

```python
theta0 = ymean-theta1*xmean
theta0
```

```python
x_test = X.sample(10)
```

```python
y_asl = y.loc[x_test.index]
```

```python
y_predicted = theta0+theta1*x_test
```

```python
pd.DataFrame({'Prognoz':y_predicted, 'savdo soni': y_asl})
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

MAE = mean_absolute_error(y_asl, y_predicted)
RMSE = np.sqrt(mean_squared_error(y_asl, y_predicted))
print(f"{MAE=}")
print(f"{RMSE=}")
```
